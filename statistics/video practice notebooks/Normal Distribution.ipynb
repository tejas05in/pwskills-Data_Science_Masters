# **`CHECK NORMAL DISTRIBUTION USING QQ PLOT = QUANTILE QUANTILE PLOT`**

```python
import scipy.stats as stat
import pylab
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset("iris")
df.head()
```

```python
import matplotlib.pyplot as plt
```

```python
#### if you want to check whether feature is guassian or normal dsitributed
#### Q-Q plot
def plot_data(df,feature):
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(df[feature],kde=True)
    plt.subplot(1,2,2)
    stat.probplot(df[feature],dist="norm",plot=pylab)
    plt.show()    
```

```python
plot_data(df,"sepal_width")
```

```python
plot_data(df,"sepal_width")
```

```python
plot_data(df,"petal_width")
```

```python
plot_data(df,"sepal_length")
```

```python
df = sns.load_dataset("tips")
df.head()
```

```python
df.info()
```

```python
for i in df.columns:
    if df[i].dtype == "float64" or df[i].dtype == "int64":
        plot_data(df,i)
```
